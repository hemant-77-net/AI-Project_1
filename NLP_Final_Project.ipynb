Step 3.3 Remove Emoticons

```python
#Emojis
emojis = {':)': 'smile', ':-)': 'smile', ';d': 'wink', ':-E': 'vampire', ':(': 'sad',
          ':-(': 'sad', ':-<': 'sad', ':P': 'raspberry', ':O': 'surprised',
          ':-@': 'shocked', ':@': 'shocked',':-$': 'confused', ':\\': 'annoyed',
          ':#': 'mute', ':X': 'mute', ':^)': 'smile', ':-&': 'confused', '$_$': 'greedy',
          '@@': 'eyeroll', ':-!': 'confused', ':-D': 'smile', ':-0': 'yell', 'O.o': 'confused',
          '<(-_-)>': 'robot', 'd[-_-]b': 'dj', ":'-)": 'sadsmile', ';)': 'wink',
          ';-)': 'wink', 'O:-)': 'angel','O*-)': 'angel','(:-D': 'gossip', '=^.^=': 'cat',';D':'bigsmile'}
```

```python
'emojis'+emojis[';D']
```

```python
def handle_emoticons(text):
  for emoji in emojis:
    text=text.replace(emoji,'emoji'+emojis[emoji])

  return text
```

```python
slang_dict={}
for i in range(len(lines)):
  slang_dict[lines[i].split('=')[0]]=lines[i].split('=')[1][:-1]
```

```python
slang_dict
```

```python
'lol! this is so funny'.split()
```

```python
def handle_slang_words(text):
  temp=[]

  for w in text.split():
    if w.upper() in slang_dict:
      temp.append(slang_dict[w.upper()])
    else:
      temp.append(w)

  return " ".join(temp)
```

```python
handle_slang_words('lol ! this is so funny')
```

```python
dataset['review']=dataset['review'].apply(lambda x: handle_slang_words(x))
```

```python
from functools import lru_cache

@lru_cache(maxsize=500000)
def remove_stopwords(text):
  text_list=text.split()
  words=stopwords.words('english')
  clean_text=[word for word in text_list if word not in words]
  return clean_text
```

##7.5 Create RNN Architecture

```python
from keras.models import Sequential
from keras.layers import Dense,Embedding,SimpleRNN,SpatialDropout1D,Bidirectional,LSTM,GRU
from keras.optimizers import Adam
from keras.regularizers import L2
```

```python
embed_dim=128
```

```python
#to detect the TPU
tpu=tf.distribute.cluster_resolver.TPUClusterResolver.connect()

#instantiate the TPU
tpu_strategy=tf.distribute.TPUStrategy(tpu)

#Create The RNN layrs
with tpu_strategy.scope():
  model=Sequential()
  model.add(Embedding(number_of_words,embed_dim,input_length=x.shape[1]))
  model.add(SpatialDropout1D(0.4))
  model.add(SimpleRNN(196,dropout=0.2,recurrent_dropout=0.2))
  model.add(Dense(2,activation='softmax',kernel_regularizer=L2(0.001)))

  model.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=0.0001),metrics=['accuracy'])
```

```python
from keras import callbacks
early_stopping=callbacks.EarlyStopping(monitor='val_los',mode='min',patience=5,restore_best_weights=True)

model.fit(x_train,y_train,epochs=20,batch_size=512,verbose=1,validation_data=(x_valid,y_valid),callbacks=[early_stopping])
```

```python
tpu=tf.distribute.cluster_resolver.TPUClusterResolver.connect()

#instantiate the TPU
tpu_strategy=tf.distribute.TPUStrategy(tpu)

#Create The RNN layrs
with tpu_strategy.scope():
  model_brnn=Sequential()
  model_brnn.add(Embedding(number_of_words,embed_dim,input_length=x.shape[1]))
  model_brnn.add(SpatialDropout1D(0.4))
  model_brnn.add(Bidirectional(SimpleRNN(196,dropout=0.2,recurrent_dropout=0.2)))
  model_brnn.add(Dense(2,activation='softmax',kernel_regularizer=L2(0.001)))

  model_brnn.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=0.0001),metrics=['accuracy'])
```

```python
from keras import callbacks
early_stopping=callbacks.EarlyStopping(monitor='val_los',mode='min',patience=5,restore_best_weights=True)

model_brnn.fit(x_train,y_train,epochs=20,batch_size=512,verbose=1,validation_data=(x_valid,y_valid),callbacks=[early_stopping])
```

```python
tpu=tf.distribute.cluster_resolver.TPUClusterResolver.connect()

#instantiate the TPU
tpu_strategy=tf.distribute.TPUStrategy(tpu)

#Create The RNN layrs
with tpu_strategy.scope():
  model_blstm=Sequential()
  model_blstm.add(Embedding(number_of_words,embed_dim,input_length=x.shape[1]))
  model_blstm.add(SpatialDropout1D(0.4))
  model_blstm.add(Bidirectional(LSTM(196,dropout=0.2,recurrent_dropout=0.2)))
  model_blstm.add(Dense(2,activation='softmax',kernel_regularizer=L2(0.001)))

  model_blstm.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=0.0001),metrics=['accuracy'])
```

```python
from keras import callbacks
early_stopping=callbacks.EarlyStopping(monitor='val_los',mode='min',patience=5,restore_best_weights=True)

model_blstm.fit(x_train,y_train,epochs=20,batch_size=512,verbose=1,validation_data=(x_valid,y_valid),callbacks=[early_stopping])
```
